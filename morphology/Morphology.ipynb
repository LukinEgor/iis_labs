```python
from pymystem3 import Mystem
```

```python
text = ''.join(open('text', 'r').readlines())
print(text)
```

```python
m = Mystem()
analize = m.analyze(text)
print(analize)
```

```python
listGr = list(map(lambda x: x['analysis'][0]['gr'].split('=')[0].split(',')[0], 
                 list(filter(lambda x: 'analysis' in x and len(x['analysis']) > 0, analize))))

dictGr = dict(map(lambda x: (x, listGr.count(x)), listGr))

russianMap = { 
    'A': 'прилагательное',
    'ADV': 'наречие',
    'ADVPRO': 'местоименное наречие',
    'ANUM': 'числительное-прилагательное',
    'APRO': 'местоимение-прилагательное',
    'COM': 'часть композита - сложного слова',
    'CONJ': 'союз',
    'INTJ': 'междометие',
    'NUM': 'числительное',
    'PART': 'частица',
    'PR': 'предлог',
    'S': 'существительное',
    'SPRO': 'местоимение-существительное',
    'V': 'глагол'
}

russianDictGr = dict(map(lambda key: (russianMap[key], dictGr[key]), dictGr.keys()))
sortDict = sorted(russianDictGr.items(), key=lambda t: -t[1])
print(sortDict)
                  
```
