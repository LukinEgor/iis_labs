```python
from pymystem3 import Mystem
```

```python
text = ''.join(open('text', 'r').readlines())
```

```python
print(text)
```

```python
m = Mystem()
lemmas = m.lemmatize(text)
print(''.join(lemmas))
```

```python
#print(m.analyze(text))
groups = ['V', 'S', 'A', 'ADV', 'SPRO', 'NUM', 'ADVPRO', 'ANUM', 'APRO']
print(m.analyze(text)[0]['analysis'][0]['gr'])
[g for g in groups if re.search(g, m.analyze(text)[0]['analysis'][0]['gr'])]
```

```python
groups = ['V', 'S', 'A', 'ADV', 'SPRO', 'NUM', 'ADVPRO', 'ANUM', 'APRO']
sortLemmas = list(filter(lambda x: 'analysis' in x 
                         and len(x['analysis']) > 0
                         and len([g for g in groups if re.search(g, x['analysis'][0]['gr'])]) > 0
                         , m.analyze(text)))
```

```python
print(sortLemmas)
```

```python
listLemmas = list(map(lambda x: x['analysis'][0]['lex'], sortLemmas))
dictLemmas = dict(map(lambda x: (x, listLemmas.count(x)), listLemmas))
sortDict = sorted(dictLemmas.items(), key=lambda t: -t[1])
print(sortDict)
```
